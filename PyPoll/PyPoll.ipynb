```python
#Import modules
import os
import csv

#Set path and open file
csvpath = os.path.join('..', 'Untitled Folder', 'election_data.csv')
           
with open(csvpath, newline='') as csvfile:
    PyPoll = csv.reader(csvfile, delimiter=',')
    cvs_header = next(PyPoll)
    
    #Create lists to store data
    total_votes = []
    all_candidates = []
    candidates = []
    Khan_count = []
    OTooley_count = []
    Li_count = []
    Correy_count = []
    votes = []
    result = []
    
    #Find total votes and candidates
    for row in PyPoll:
        total_votes.append(row[0])
        total_number_votes = len(total_votes)
    
        all_candidates.append(row[2])
    
    #Find all unique candidates
    for x in set(all_candidates):
        candidates.append(x)
    
    #Find each candidates total vote count and percentage
    for candidate in all_candidates:
        if candidate == "Khan":
            Khan_count.append(all_candidates)
            Khan = len(Khan_count)
            K_percent = '{0:.3f}'.format((Khan / total_number_votes) * 100)
        elif candidate == "O'Tooley":
            OTooley_count.append(all_candidates)
            OTooley = len(OTooley_count)
            O_percent = '{0:.3f}'.format((OTooley / total_number_votes) * 100)
        elif candidate == "Li":
            Li_count.append(all_candidates)
            Li = len(Li_count)
            L_percent = '{0:.3f}'.format((Li / total_number_votes) * 100)
        else:
            Correy_count.append(all_candidates)
            Correy = len(Correy_count)
            C_percent = '{0:.3f}'.format((Correy / total_number_votes) * 100)
    
    #Find the winner
    for candidate in candidates:
        votes.append(int(all_candidates.count(candidate)))
    
    results = {candidates[i]: votes[i] for i in range(len(votes))}
    
    winner = max(results, key=results.get)
    
    #Print the Analysis
    print(f"Election Results")
    print(f"-------------------------")
    print(f"Total Votes: {total_number_votes}")
    print(f"-------------------------")
    print(f"Khan: {K_percent}% ({Khan})")
    print(f"Correy: {C_percent}% ({Correy})")
    print(f"Li: {L_percent}% ({Li})")
    print(f"O'Tooley: {O_percent}% ({OTooley})")
    print(f"-------------------------")
    print(f"Winner: {winner}")
    print(f"-------------------------")
    
#Export to Text File
output_path = os.path.join("..", "Untitled Folder", "Election_Analysis.txt")

with open(output_path, 'w') as text:
    text.write(f"Election Results\n")
    text.write(f"-------------------------\n")
    text.write(f"Total Votes: {total_number_votes}\n")
    text.write(f"-------------------------\n")
    text.write(f"Khan: {K_percent}% ({Khan})\n")
    text.write(f"Correy: {C_percent}% ({Correy})\n")
    text.write(f"Li: {L_percent}% ({Li})\n")
    text.write(f"O'Tooley: {O_percent}% ({OTooley})\n")
    text.write(f"-------------------------\n")
    text.write(f"Winner: {winner}\n")
    text.write(f"-------------------------\n")
```
